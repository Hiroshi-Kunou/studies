```python
dataset3=randn(25)
sns.rugplot(dataset3)
plt.ylim(0,1)
```

```python
plt.hist(dataset3,alpha=0.3)
sns.rugplot(dataset3)
```

```python
sns.rugplot(dataset3);

x_min=dataset3.min()-2
x_max=dataset3.max()+2

x_axis=np.linspace(x_min,x_max,100)
bandwidth=((4*dataset3.std()**5)/(3*len(dataset3)))**.2

kernel_list=[]

for data_point in dataset3:
    
    kernel=stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    kernel=kernel/kernel.max()
    kernel=kernel*.4
    plt.plot(x_axis,kernel,color='gray',alpha=0.5)
    
plt.ylim(0,1)
```

```python
sum_of_kde=np.sum(kernel_list,axis=0)
fig=plt.plot(x_axis,sum_of_kde,color='indianred')
sns.rugplot(dataset3,c='indianred')
plt.yticks([])
plt.suptitle("Sum of the Basis Functions")
```

```python
sns.kdeplot(dataset3)
```

```python
sns.rugplot(dataset3,color='black')
for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset3,bw=bw,label=bw)
```

```python
kernel_options=["biw","cos","epa","gau","tri","triw"]
for kern in kernel_options:
    sns.kdeplot(dataset3,kernel=kern,label=kern)
```

```python
for kern in kernel_options:
    sns.kdeplot(dataset3,kernel=kern,label=kern,shade=True,alpha=0.5)
```

```python
sns.kdeplot(dataset3,vertical=True)
```
